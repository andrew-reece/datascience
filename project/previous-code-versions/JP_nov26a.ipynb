```python

## NYT API keys
api_key_article = "851e7d0a131bee9bc01097470c238637:13:47475506"
api_key_community = "519167db119ee6408c4ee51b3c391e11:0:47475506"
api_key_geo = "a984ad78bf017f0ade1fcd980aa6353f:15:47475506"
api_key_popular = "09dfaf288ad6c2ec46893a27ca758d41:19:47475506"
api_key_movies = "e8a48f7d7731698b05267146c681c352:5:47475506"
api_key_semantic = "9063b41607bbf486247b8e596a1456b8:7:47475506"
api_key_newswire = "209ebb7b0ab44094970e8b39c63fea7e:2:47475506"
api_key_timestags = "43b3366f288db10cb019fd532299723f:10:47475506"


begindate = "20000101" #YYYYMMDD
enddate = "20131112" #YYYYMMDD


## just picked a few terms meant to have a fair spread in content
terms = ['mindfulness', 'debt', 'kardashian', 'obama', 'romney', 'obamacare', 'god', 'terrorism', 'nazi', 'hussein', 'war', 'depression', 'abortion', 'sex']

# intially an empty dataframe
data_df = pd.DataFrame()

## loop through terms
for term in terms:
    
    ## api request for each term
    url = ''.join(["http://api.nytimes.com/svc/search/v2/articlesearch.json?q=", term,
                   "&begin_date=", begindate, 
                   "&end_date=", enddate,
                   "&api-key=",api_key_article])
                   
    req = requests.get(url).text
    
    ## decode into json dicts
    jsons = json.loads(req)
    
    ## loop through each article returned from API request
    for doc in jsons['response']['docs']:

        # making dataframe; a term will appear in multiple rows
        doc_df = pd.DataFrame([term], columns = ['term'])
        
        # alternative way of dealing with none-types
        if json.dumps(doc['abstract']) != 'null':
            
            # encode weird ascii stuff
            abstract = [doc['abstract'].encode('utf-8')]
            
            # making dataframe and adding abstracts
            doc_df['abstract'] = abstract
            
        # this way we take all data, and we can do away with what we don't want later    
        else:
            # to avoid complication, just use the STRING 'NaN'
            abstract = 'NaN'
            doc_df['abstract'] = abstract
            
        # can add an if-else clause for anything you want to get, then add to doc_df
        if json.dumps(doc['lead_paragraph']) != 'null':
            
            lead = [doc['lead_paragraph'].encode('utf-8')]
            doc_df['lead'] = lead
            
        else:
            lead = 'NaN'
            doc_df['lead'] = lead
        
        data_df = pd.concat([data_df, doc_df]).reset_index(drop=True)
        
        
print len(data_df)
print len(data_df[(data_df.lead != 'NaN') & (data_df.abstract != 'NaN')])
print len(data_df[data_df.lead != 'NaN'])
data_df.lead[data_df.lead != 'NaN']
```

```python
"""
Function
--------
calibration_plot
 
Builds a plot like the one above, from a classifier and review data
 
Inputs
-------
clf : Classifier object
    A MultinomialNB classifier
X : (Nexample, Nfeature) array
    The bag-of-words data
Y : (Nexample) integer array
    1 if a review is Fresh
"""   
 
from scipy import *
from matplotlib.ticker import FormatStrFormatter
import math
 
#define function to plot calibration graphs for given clf and bag of word x,y
def calibration_plot(clf, X, Y):
    
    # setting number of bins
    binnumber = 20
    
    # get expected P(unhappy),P(happy)
    probabilities = clf.predict_proba(X)
    
    #get the P(happy) column
    probabilities = probabilities[:,1]
    
    #get hist (bin counts) and edges of the histogram
    hist, edges = np.histogram(probabilities, bins=binnumber)
    
    # get bin number for each value according to the assumed histogram 
    # subtract 1 from edges because it is the cumulation of 1
    binnrcontent = np.digitize(probabilities, edges[:-1], right = False)
    
    # set bincontent to data (probabilities of p(happy)
    bincontent = probabilities
    
    # create dataframe with three columns: bin (binnr), content (bincontent), 
    # and realhappy (actual evaluation of happiness)   
    myframe = pd.DataFrame({'bin': binnrcontent, 'content': bincontent, 'realhappy': Y})
    myframe = myframe.sort('bin').reset_index(drop=True)
        
    complete_frame = myframe.copy()
        
    # adding bin numbers which were excluded by np.digitize
    for i in xrange(binnumber):
        
        if not myframe.bin.apply(lambda x: x == i+1).any():
        
            extra_bin = pd.DataFrame({'bin': int(i+1), 'content': np.nan , 'realhappy': np.nan}, index = [0])
            complete_frame = pd.concat([complete_frame, extra_bin]).reset_index(drop=True)
    
    #code section to get fractions of bins (y-axis)
    #create column bincount to store the total count of samples in the respective bin (stored redundantly)
    myframe['bincount'] = hist[myframe['bin']-1] 
    
    complete_frame['bincount'] = hist[complete_frame['bin']-1] 
    
    #print myframe.bincount.values
    
    #create column to calculate fraction of realfresh divided by the respective bin total count (done for every row!)
    
    myframe['fraction'] = 0.
    myframe['fraction'][myframe.bincount != 0] = myframe['realhappy'][myframe.bincount != 0]*1.0 / myframe['bincount'][myframe.bincount != 0]
    
    complete_frame['fraction'] = np.nan
    complete_frame['fraction'][complete_frame.bincount != 0] = complete_frame['realhappy'][complete_frame.bincount != 0]*1.0 / complete_frame['bincount'][complete_frame.bincount != 0]
    
    #get fraction of bins by grouping by bin and sum the fraction in aggregation
    fractions = pd.DataFrame(myframe.groupby('bin').fraction.sum(), columns = ['happy'])
    
    complete_fractions = pd.DataFrame(complete_frame.groupby('bin').fraction.sum(), columns = ['happy'])
    
    #turn results into list for easy usage in the plot code
    fractionlist = list(fractions['happy'])
    fractioncomplete = list(complete_fractions['happy'])
    
    
    ### code section to get bin values (x-axis, 20 in numbers) ###
    
    #get means of bins by grouping by bin and aggregated mean function of content column
    contents = pd.DataFrame(myframe.groupby('bin').content.mean(), columns = ['content'])
    
    complete_contents = pd.DataFrame(complete_frame.groupby('bin').content.mean(), columns = ['content'])
    
    #turn results into list for easy usage in the plot code
    contentslist = list(contents['content'])
   
    contentscomplete = list(complete_contents['content'])
    
    #calculate errorbar values for each bin (using fractions that is grouped by bins)
    fractions['bincount'] = hist[hist != 0]
    
    fractions['happy'] = fractions['happy'].apply(lambda x: float(x))
    
    fractions['error'] = fractions.apply(lambda x: 
                    sqrt((x['happy'] * (1 - x['happy'])) / x['bincount']), axis=1)
    
    #turn results into list for easy usage in the plot code
    errorlist = list(fractions['error'])   
    

    ### plot section ###
    
    # tell plt to use figure 1 because more than one plot is drawn   
    plt.figure(1)
    
    # set x: the mean values of the bins
    x = contentslist
    
    # set y: the calculated fractions per bin
    y = fractionlist
    
    # print x -> f(x) with red dots
    plt.plot(x, y, 'ro-', label='expected P(happy)/observed happiness fraction')
    
    # calculate trend line and make line function
    line = np.polyfit(x,y,1)
    func = np.poly1d(line)
    
    # set trendlinetext for legend
    trendlinetxt = 'Trendline: %sx + (%s)' % (str(round(line[0],2)), 
                                              str(round(line[1],2)))
    plt.plot(x, func(x), '--', color = 'blue', label=trendlinetxt)
   
    # create second line for reference (y=x)
    line = np.polyfit(x,x,1)
    # convert into function
    func = np.poly1d(line)
    
    # set trendlinetext for legend
    trendlinetxt = 'Y=X reference line: %sx + (%s)' % (str(round(line[0],2)), 
                                                       str(round(line[1],2)))
    plt.plot(x, func(x), 'k', color = 'black', label=trendlinetxt)
    plt.errorbar(x, y, yerr=errorlist, fmt='o', label = 'Estimated Uncertainty')
   
    # making labels
    plt.legend(loc = 'lower right', prop={'size':10})
    plt.xlabel("Expected P(Happy)")
    plt.ylabel("Observed Happiness Fraction")
    plt.show()
   
    # prepare plt to plot different plot
    fig, ax = plt.subplots()
    # calc histogram
    counts, bins, patches = ax.hist(probabilities, edgecolor='black', 
            bins = binnumber, label='happiness/#reviews - (bincount/happiness fract)')
    
    # get bin centers
    bin_centers = 0.5 * np.diff(bins) + bins[:-1]
    
    # iterate trough fractionvals and bincounts
    xlabellist = []
    for i in xrange(len(fractioncomplete)):
        
        if not np.isnan(fractioncomplete[i]):
            perc = '%0.0f%%' % (fractioncomplete[i] * 100)
            bincount = '%0.0f' % (counts[i])
            probs = '%0.0f' % (contentscomplete[i])
            storestr = '(%s / %s) - %s' % (bincount,perc,probs)
            xlabellist.append(storestr)
        else:
            xlabellist.append("")
        
    # set the label position based on bin_centers
    ax.set_xticks(bin_centers)
    ax.set_xticklabels(xlabellist,rotation=45, rotation_mode="anchor", ha="right", size=10)
    
    # get more bottom of the plot
    plt.subplots_adjust(bottom=0.15)
   
    # making labels
    plt.legend(loc = 'upper left', prop={'size':16})
    plt.xlabel("Happiness (1/0 Happy/Unhappy)")
    plt.ylabel("Number of Reviews")
    plt.show()

 
# test function
calibration_plot(clf, Xtest, Ytest)

```
