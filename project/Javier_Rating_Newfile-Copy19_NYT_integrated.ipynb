#### 2.2 Stanford MovieDB (stanford_df) <a name="sf_make_data"></a>
25k reviews divided in 12.5k positive ones and 12.5k negative. Ratings range from 1-10 for 1 negative and 10 positive. All neutral reviews have been excluded beforehand. Dataframe will be constructed from filesystem structure. 

Set up function to create stanford dataframe. A function is used to create reusability for the later use in case of testing data.

```python
# here the parameter "data_set" is a string (either 'train' or 'test')
def make_stanford_df(data_set):

    # making lists of negative and positive training data filenames, respectively
    path_neg = "stanford_movie/%s/neg/" % data_set
    neg_list = os.listdir(path_neg)
    
    path_pos = "stanford_movie/%s/pos/" % data_set
    pos_list = os.listdir(path_pos)
    
    # making sure we're only getting the filenames we want
    # there seems to be hidden files at least in the neg folder
    neg_list = [i for i in neg_list if i[0].isdigit()]
    pos_list = [i for i in pos_list if i[0].isdigit()]
    
    # making dataframe for negative training data
    neg_df = pd.DataFrame(neg_list, columns = ['filename'])
    
    neg_df['id'] = neg_df.apply(lambda x: 
                            'neg' + x['filename'][:x['filename'].index('_')], axis=1)
    
    neg_df['rating'] = neg_df.apply(lambda x: 
                            x['filename'][x['filename'].index('_')+1:-4], axis=1)
    
    neg_df['path'] = path_neg
    
    # making dataframe for positive training data
    pos_df = pd.DataFrame(pos_list, columns = ['filename'])
    pos_df['id'] = pos_df.apply(lambda x: 
                            'pos' + x['filename'][:x['filename'].index('_')], axis=1)
    pos_df['rating'] = pos_df.apply(lambda x: 
                            x['filename'][x['filename'].index('_')+1:-4], axis=1)
    
    pos_df['path'] = path_pos
    
    # concatenating negative and positive training data into one dataframe
    stanford_df = pd.concat([neg_df, pos_df], ignore_index=True)
    
    # adding column with the movie reviews
    stanford_df['review'] = stanford_df.apply(lambda x: 
                            open(x['path'] + x['filename'], 'r').read(), axis=1)

    return stanford_df
```

```python
def call_nyt_api(api, url_base, api_key, range_min, range_max, increment):

    dicts = []
    
    ## range_min, range_max, increment, page_param and offset determine how often we request new pages of api calls
    ## they're called different things in the two APIs, thus all the hoopla
    ## we basically want to have about 100 unique articles in each category
    
    page_param = "&offset=" if api == "POPULAR" else "&page="
    
    for offset in range(range_min, range_max, increment):

        print 'running range:',offset,'to',offset+increment,'\n'
        
        url = ''.join([url_base,
                        api_key,
                        page_param,
                        str(offset)])
        ## print url
        
## get raw html for each article
        
        api_response = requests.get(url).text
        
        jsons = json.loads(api_response)

        try:
            json_dict = jsons["results"] if api == 'POPULAR' else jsons["response"]["docs"]
        except:
            continue
            
        dicts.append(json_dict)    
        
    return dicts
```

```python
def make_soup(url):
    
## load raw html from url

    raw_article_html = requests.get(url).text
     
## create soup object for parsing

    soup = BeautifulSoup(raw_article_html)
    
    return soup
    
```

```python
def get_article_title(json, api):
    
    title = json["title"] if api == 'POPULAR' else json["headline"]["main"]
    
    return title
```

```python
def get_article_metadata(soup):
            
## after looking through the actual article page source code
## we can see that the unique article id is stored in a meta tag:
## <meta content="100000002558432" itemprop="identifier" name="articleid"/>

## there's also some other metadata we're interested in:
## eg. <meta name="PT" content="Article">

## so we parse out the content of these tags in order to get relevant metadata

## first get content type and genre (eg. 'Article / News')

## we make each a try/except block to prevent random errors that sometimes do show up

    try:
        content_type = soup.find("meta", attrs={"name":"PT"}).attrs["content"]
    except:
        content_type = ''
    try:
        content_genre = soup.find("meta", attrs={"name":"PST"}).attrs["content"]
    except:
        content_genre = ''
    try:
        author = soup.find("meta", attrs={"name":"author"}).attrs["content"]
    except:
        author = ''
    try:
        page_num = soup.find("span", attrs={"itemprop":"printPage"}).string
    except:
        page_num = ''
        
        ## NB: on some blogs, the page number is in a block of text, and we'd need to do regex extraction to get it.
        ##
        ## eg. http://well.blogs.nytimes.com/2013/11/11/women-find-orgasms-elusive-in-hookups/
        ## at the end of the actual article text there's this <h6> tag:
        ## <h6 class="metaFootnote">A version of this article appears in print on 11/12/2013, on page D1 of the NewYork edition with the headline: In Hookups, Inequality Still Reigns.</h6>
   
    try:
        category = soup.find("meta", attrs={"name":"CG"}).attrs["content"]
    except:
        category = ''
    try:
        sub_category = soup.find("meta", attrs={"name":"SCG"}).attrs["content"]
    except:
        sub_category = ''
        
    if content_type == "Blogs":
        
        id_tag = soup.find("meta", attrs={"name":"communityMD5"})
        
    else:
        
        id_tag = soup.find("meta", attrs={"itemprop": "identifier", "name":"articleid"})

        
    article_id = id_tag.attrs["content"] if (id_tag) else ''

## we don't want to include weird content (funky multimedia, or obituaries, for instance)

    include = False if not ((content_type == "Blogs") or (content_type == "Article")) or ((content_genre == "Obituary (Obit)") or (content_genre == "Schedule")) else True
        
        
    return article_id, content_type, content_genre, category, sub_category, author, page_num, include 
```

```python
##
## get article text
##


## helper function for beautiful soup
##
def is_p_and_has_no_class(tag):
    
    return (not tag.has_attr('class')) and (tag.name=='p')



## main function
##
def get_article_text(soup, content_type, genre):

    article_body = []
        
    ## after looking through the actual article page source code
    ## we can see that this 'itemprop' attribute stores all blocks of article text
    ## so we parse out only these blocks from the whole text dump (there are lots of ads, images, menus, etc)
    
    ## text scraping requires finding the different tags that mark article body text in each section
    #  eg. slightly different with blogs & reviews than with regular news 

    if content_type == "Blogs":
    
        text = soup.findAll(is_p_and_has_no_class)

    elif genre == "Review":
        
        text = soup.findAll(itemprop="reviewBody")
        
    else:
        
        text = soup.findAll("p", attrs={"itemprop":"articleBody"})
    
    ## testing
    ## print 'this is the article text:',text
    
    return text
```

```python
def cleanup_text(text, content_type):
    
## apostrophes (') get turned into a funny French a
## so we sub back in ' with regex
## we also get rid of carriage returns and other useless whitespace 

    remove_a = re.compile(u'â')
    remove_w = re.compile('[\n\r\t]')  
    remove_comma = re.compile(',')
    
    ptags = []

    try:   
        
        ct = 0   ## this counter is for error handling
        for t in text:
            
            ct+=1
            
            try:
                ptags.append(t.contents[0].string)
                
            except:
                # print 'failed at:',ct    ## some blogs barf on the <p> html structure for some reason
                break
   
        ptags = [ remove_a.sub(r"", p) for p in ptags ]

        ptags = [ remove_w.sub(r"", p) for p in ptags ]

        ptags = [ remove_comma.sub(r"", p) for p in ptags ]

        cleaned_text = ' '.join(ptags)
        
    except:
        # print text
        cleaned_text = 'error_text'
        
    return cleaned_text
```

```python
## extract article body text from raw html

def extract_articles(api, jsons, target_count=0):

## the API return only gives us metadata and lead paragraph
## but we do get the URL
## so here we load each article URL and scrape the text
## then we store the actual article text bodies in a list

    articles = []
    
    popular = 1 if api == "POPULAR" else 0
    
    ## counter is for matching roughly the same number of popular and non-popular articles
    ## we do popular articles first.  so when collecting unpopular articles, we keep a running count,
    ## and when we match the number of popular articles already collected, we can stop scraping.
    counter = 0
    
## STEPS IN THIS CODE BLOCK:
##
## 1) loop through each article's json
## 2) scrape article text from url
## 3) grab metadata
## 4) clean article text
## 5) store in master list of articles

## NB: The Article and Most Popular APIs have different json trees, 
## so at times we need conditional branches to assign the correct paths to the nodes we want
    
    #1
    
    for json_set in jsons:
        
        for j in json_set:
         
            if not popular:
                
                if counter >= target_count:
                    break
                else:    
                    counter += 1
            
        #2
            url = j["url"] if api == 'POPULAR' else j["web_url"]
            
            ## print url
            
            soup = make_soup(url) 

                
        #4
            title = get_article_title(j, api)
            
            ## print title
            
            ## NB:
            ## since it's December, 
            ## there are a few 'best of 2013' articles and reviews which are really mostly blurbs and links
            ## this search pattern flags those articles for exclusion later
            
            its_a_survey = re.search('of 2013$', title)
            
            survey = True if its_a_survey else False
            
            
            ## get section name
            ## eg. World, U.S., Books, Science
            
            section_formal = j["section"] if api == 'POPULAR' else j["section_name"]
             
            ## get other metadata
            ## eg. author name, print page #, etc.
            
            article_id, content_type, content_genre, section, sub_category, author, page_num, include = get_article_metadata(soup)
            
            nyt_original = False if author == "THE ASSOCIATED PRESS" else True
            
        #3
            text = get_article_text(soup, content_type, content_genre)
            
        
        ## some nodes fail on the regex in our cleanup function, so we do try/except
                
            try:
                    
                cleaned_text = cleanup_text(text, content_type)
                    
            except: 
                    
                print 'error with regex for article:', title
                cleaned_text = 'text cleaning error'
                continue
                
        ## NYT actually gives word_length metadata, but only on the Article API, not on Most Popular
        ## so if NYT doesn't give it to us, we have a function that approximates word count
        
            try:    
                
                word_count = j["word_count"]
                
            except KeyError:
                
                word_count = ''
                
            
            ## append article text to master list 'articles'
                
        #5
        
            ## exclusion criteria:
            ##
            ## include:         see get_article_metadata() comments
            ##
            ## word_count > 20: sometimes articles are actually just editorial cartoons with a caption.
            ##                  no real article has <20 words, and no cartoon captions seem to have >20 words.
            ##
            ## survey:          see NB above in this block which starts "since it's December..."
            ## 
            ## nyt_original:    some articles are from AP. 
            ##                  this excludes based on author byline, but there's a metadata tag for this as well
            
            if (include) and (word_count > 20) and (not survey) and nyt_original:
                    
                articles.append([article_id, 
                                 popular, 
                                 content_type, 
                                 content_genre,
                                 section_formal, 
                                 section,
                                 sub_category,
                                 title,
                                 author,
                                 word_count,
                                 url,
                                 page_num,  
                                 cleaned_text])

    return articles
```

```python
def make_article_df(articles):
    
    df = pd.DataFrame(index = range(0,len(articles)), 
                              columns = ["article_id",
                                         "popular",
                                         "content_type",
                                         "genre",
                                         "section_formal",
                                         "section",
                                         "sub-category",
                                         "title",
                                         "author",
                                         "post_length",
                                         "url",
                                         "page_num",
                                         "text"])

    for idx, a in enumerate(articles):
    
        df.ix[idx, "article_id"] = a[0]
        df.ix[idx, "popular"] = a[1]
        df.ix[idx, "content_type"] = a[2]
        df.ix[idx, "genre"] = a[3]
        df.ix[idx, "section_formal"] = a[4]
        df.ix[idx, "section"] = a[5]
        df.ix[idx, "sub-category"] = a[6]
        df.ix[idx, "title"] = a[7]
        df.ix[idx, "author"] = a[8]
        df.ix[idx, "post_length"] = a[9]
        df.ix[idx, "url"] = a[10]
        df.ix[idx, "page_num"] = a[11]
        df.ix[idx, "text"] = a[12]
        
    return df
```

```python
def get_date_range():
   
    today = time.strftime("%Y%m%d", time.localtime())
    
    days_to_subtract = 30
    
    one_month_ago = date.today() - timedelta(days=days_to_subtract)

    one_month_ago = one_month_ago.strftime("%Y%m%d")

    return today, one_month_ago
```

#### Make XY Function

This function takes in the prepared dataframe source and outputs X (an array of features) and Y (an array of known scores).

```python
# defining the make_xy() function
# "rating_col" parameter is the name of the column where the ratings are;
# can be continuous or binary values
# features parameter is a list of the feature functions
def make_xy(prep_data, rating_col, features):
    
    # using copy of dataframe just in case
    prep_copy = prep_data.copy()
    
    # converting relevant columns to floats
    
    #map(features, lambda y:  prep_copy[y.__name__] = prep_copy[y.__name__].apply(lambda x: float(x)))
    
    for func in features:
        #print func
        prep_copy[func.__name__] = prep_copy[func.__name__].apply(lambda x: float(x))
        prep_copy = prep_copy[~prep_copy[func.__name__].isnull()].reset_index(drop = True)
    
    feature_array = []
    for i in xrange(len(prep_copy)):
        feature_list = [float(prep_copy[func.__name__][prep_copy.index == i].values[0]) for func in features]
        feature_array.append(feature_list)
    
    #map(features, lambda x: df_copy[x.__name__] = np.nan)
    #map(features, lambda y: df_copy[y.__name__] = df_copy.text_col.apply(lambda x: y(x)))
    
    # making 2D array of features
    x = np.array(feature_array)
    
    # renaming the column that contains the ratings
    prep_copy = prep_copy.rename(columns={rating_col: 'rating_col'})
    
    # making numpy array of the naive happy values we made up
    y = np.array(prep_copy.rating_col.values)
    
    # returning x and y in a tuple
    return x, y
```

```python

### TESTING COMBINATIONS OF FEATURES
###
###

def test_features(train_df, test_df, all_features, min_max_scale, min_max_emp, output, MLmodel):
    start = time.time()
    #print 'start \n', start
    all_features_str = [i.__name__ for i in all_features]
    
    ### need to reorder features in dataframes to math all_features
    ## training data
    train_df_copy = train_df.copy()
    train_df_copy = train_df_copy.rename(columns={'FK_GradeLevel': 'fk_gradelevel'}) ##need to sort to make lower case
    
    # pandas indexing is INCLUSIVE unlike normal indexing
    nonfeatures_train_df = train_df_copy.ix[:,'text']
    features_train_df = train_df_copy.ix[:, :-1]
    sorted_features_train_df = features_train_df.sort_index(axis=1)
    sorted_full_train_df = sorted_features_train_df.join(nonfeatures_train_df)

    ## repeating same process above. testing data
    test_df_copy = test_df.copy()
    test_df_copy = test_df_copy.rename(columns={'FK_GradeLevel': 'fk_gradelevel'})
    
    # pandas indexing is INCLUSIVE unlike normal indexing
    nonfeatures_test_df = test_df_copy.ix[:, 'text']
    features_test_df = test_df_copy.ix[:, :-1]
    sorted_features_test_df = features_test_df.sort_index(axis=1)
    sorted_full_test_df = sorted_features_test_df.join(nonfeatures_test_df)
        
    xtrain, ytrain = make_xy(train_df, output, all_features)
    xtest, ytest = make_xy(test_df, output, all_features)

    feature_nums = list(np.arange(len(all_features)))
    
    num_combo_list = []
    combo_list = []
    combo_list_str = []

    for i in xrange(len(all_features)):
        
        # getting a list of all possible feature/function combinations
        combo = list(combinations(all_features, i+1))
        combo_str = list(combinations(all_features_str, i+1))
        combo_list.extend(combo)
        combo_list_str.extend(combo_str)
        
        # making corresponding combinations of numbers
        # these will match the order of the previous combos and will correspond
        # to the column indices of the features in the modified df
        
        num_combo = list(combinations(feature_nums, i+1))
        num_combo_list.extend(num_combo) #combinations that correspond 

    combo_list = [list(i) for i in combo_list]
    combo_list_str = [list(i) for i in combo_list_str]
    num_combo_list = [list(i) for i in num_combo_list]
    
    combo_list_str = np.array(combo_list_str)
    combo_df = pd.DataFrame(combo_list_str, columns=['combo_features'])
    
    #print combo_df.combo_features
    
    actuallist=[]
    predictedlist=[]
    differencelist=[]
    accuracylist1=[]
    accuracylist2=[]
    stdlist=[]
    
    bestpredictedvaluelist = []
    bestrealvaluelist=[]
    bestpredictedtextlist=[]
    worstpredictedvaluelist = []
    worstrealvaluelist=[]
    worstpredictedtextlist=[]
    
    
    xtraindf = pd.DataFrame(xtrain)
    
    #print '\n', xtraindf
    
    xtestdf = pd.DataFrame(xtest)
    ytestdf = pd.DataFrame(ytest)
    
    for i, val in enumerate(combo_list):
        now = time.time()
        #print 'now:', now
        print 'combo number', i
        print 'time elapsed', float(now) - float(start)
        
        x_indices = num_combo_list[i]
        
        xtrainarray = np.array(xtraindf[x_indices])
        xtestarray = np.array(xtestdf[x_indices])
        
        clf = MLmodel
        clf.fit(xtrainarray, ytrain)
        
        #sum_df = pd.DataFrame(clf.predict(xtestarray))
        #sum_df['actual'] = ytest 
        
        predicted = clf.predict(xtestarray)
        actual = ytest
        difference = abs(actual-predicted)
        
        #print 'diff', type(difference)

        #print 'mean', np.mean(difference, dtype=float)

        
        rangeval = abs(min_max_scale[1] - min_max_scale[0])        
        minmaxval = abs(min_max_emp[1] - min_max_emp[0])
        
        accuracy1 = (100. - np.mean(difference, dtype=float) / rangeval * 100.)
        accuracy2 = (100. - np.mean(difference, dtype=float) / float(minmaxval) * 100.)
        
        std = np.std(difference, dtype=float)
        
        bestindex = np.where(difference == min(abs(difference)))
        bestrealvalue = actual[bestindex]
        bestpredicted = predicted[bestindex]
        bestpredictedtext = test_df.irow(bestindex).text
        
        worstindex = np.where(difference == max(abs(difference)))
        worstrealvalue = actual[np.where(difference==max(abs(difference)))]
        worstpredicted = predicted[worstindex]
        worstpredictedtext = test_df.irow(worstindex).text

        actuallist.append(actual)
        predictedlist.append(predicted)
        differencelist.append(difference.astype(float))
        accuracylist1.append(accuracy1)
        accuracylist2.append(accuracy2)
        
        stdlist.append(std)
        
        bestpredictedvaluelist.append(bestpredicted)
        bestrealvaluelist.append(bestrealvalue)
        bestpredictedtextlist.append(bestpredictedtext)  
        
        worstpredictedvaluelist.append(worstpredicted)
        worstrealvaluelist.append(worstrealvalue)
        worstpredictedtextlist.append(worstpredictedtext)
        
    combo_df['actual'] = actuallist
    combo_df['predicted'] = predictedlist
    combo_df['difference'] = differencelist
    
    combo_df['accuracy_scale'] = accuracylist1
    combo_df['accuracy_emp'] = accuracylist2
    
    combo_df['std'] = stdlist
    
    combo_df['bestpredictedvalue'] = bestpredictedvaluelist
    combo_df['bestpredictedrealvalue'] = bestrealvaluelist
    combo_df['bestpredictedtext'] = bestpredictedtextlist
    
    combo_df['worstpredictedvalue'] = worstpredictedvaluelist
    combo_df['worstpredictedrealvalue'] = worstrealvaluelist
    combo_df['worstpredictedtext'] = worstpredictedtextlist 
    
    return combo_df
```
