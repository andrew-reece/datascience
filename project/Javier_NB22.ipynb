#### 2.2 Stanford MovieDB (stanford_df) <a name="sf_make_data"></a>
25k reviews divided into 12.5k positive ones and 12.5k negative. Ratings range from 1-10 (1 being negative and 10 being positive). All neutral reviews have been excluded beforehand. Dataframe will be constructed from filesystem structure. 

Set up function to create stanford dataframe. A function is used to create reusability for the later use in case of testing data.

```python
# here the parameter "data_set" is a string (either 'train' or 'test')
def make_stanford_df(data_set):

    # making lists of negative and positive training data filenames, respectively
    path_neg = "stanford_movie/%s/neg/" % data_set
    neg_list = os.listdir(path_neg)
    
    path_pos = "stanford_movie/%s/pos/" % data_set
    pos_list = os.listdir(path_pos)
    
    # making sure we're only getting the filenames we want
    # there seems to be hidden files at least in the neg folder
    neg_list = [i for i in neg_list if i[0].isdigit()]
    pos_list = [i for i in pos_list if i[0].isdigit()]
    
    # making dataframe for negative training data
    neg_df = pd.DataFrame(neg_list, columns = ['filename'])
    
    neg_df['id'] = neg_df.apply(lambda x: 
                            'neg' + x['filename'][:x['filename'].index('_')], axis=1)
    
    neg_df['rating'] = neg_df.apply(lambda x: 
                            x['filename'][x['filename'].index('_')+1:-4], axis=1)
    
    neg_df['path'] = path_neg
    
    # making dataframe for positive training data
    pos_df = pd.DataFrame(pos_list, columns = ['filename'])
    pos_df['id'] = pos_df.apply(lambda x: 
                            'pos' + x['filename'][:x['filename'].index('_')], axis=1)
    pos_df['rating'] = pos_df.apply(lambda x: 
                            x['filename'][x['filename'].index('_')+1:-4], axis=1)
    
    pos_df['path'] = path_pos
    
    # concatenating negative and positive training data into one dataframe
    stanford_df = pd.concat([neg_df, pos_df], ignore_index=True)
    
    # adding column with the movie reviews
    stanford_df['review'] = stanford_df.apply(lambda x: 
                            open(x['path'] + x['filename'], 'r').read(), axis=1)

    return stanford_df
```

```python
def call_nyt_api(api, url_base, api_key, range_min, range_max, increment):

    dicts = []
    
    ## range_min, range_max, increment, page_param and offset determine how often we request new pages of api calls
    ## they're called different things in the two APIs, thus all the hoopla
    ## we basically want to have about 100 unique articles in each category
    
    page_param = "&offset=" if api == "POPULAR" else "&page="
    
    for offset in range(range_min, range_max, increment):

        print 'running range:',offset,'to',offset+increment,'\n'
        
        url = ''.join([url_base,
                        api_key,
                        page_param,
                        str(offset)])
        ## print url
        
## get raw html for each article
        
        api_response = requests.get(url).text
        
        try:
            jsons = json.loads(api_response)
        except:
            continue
            
        try:
            json_dict = jsons["results"] if api == 'POPULAR' else jsons["response"]["docs"]
        except:
            continue
            
        dicts.append(json_dict)    
        
    return dicts
```

```python
def make_soup(url):
    
## load raw html from url

    raw_article_html = requests.get(url).text
     
## create soup object for parsing

    soup = BeautifulSoup(raw_article_html)
    
    return soup
    
```

```python
def get_article_title(json, api):
    
    title = json["title"] if api == 'POPULAR' else json["headline"]["main"]
    
    return title
```

```python
def get_article_metadata(soup, j, api):
   
    ## get title
    
    title = get_article_title(j, api)
            
    ## print title
    
    ## NB:
    ## since it's December, 
    ## there are a few 'best of 2013' articles and reviews which are really mostly blurbs and links
    ## this search pattern flags those articles for exclusion later
    
    its_a_survey = re.search('of 2013$', title)
    
    survey = True if its_a_survey else False
    
    ## get section name
    ## eg. World, U.S., Books, Science
    
    section_formal = j["section"] if api == 'POPULAR' else j["section_name"]
    
    ## an exclusion criterion...'Corrections' means amendments to previous articles, they're not really articles.
    
    corrections = True if section_formal == "Corrections" else False
    
## after looking through the actual article page source code
## we can see that the unique article id is stored in a meta tag:
## <meta content="100000002558432" itemprop="identifier" name="articleid"/>

## there's also some other metadata we're interested in:
## eg. <meta name="PT" content="Article">

## so we parse out the content of these tags in order to get relevant metadata

## first get content type and genre (eg. 'Article / News')

## we make each a try/except block to prevent random errors that sometimes do show up

    try:
        content_type = soup.find("meta", attrs={"name":"PT"}).attrs["content"]
    except:
        content_type = ''
        
        
    try:
        content_genre = soup.find("meta", attrs={"name":"PST"}).attrs["content"]
    except:
        content_genre = ''
        
        
    try:
        author = soup.find("meta", attrs={"name":"author"}).attrs["content"]
    except:
        author = ''
    
    nyt_original = False if author == "THE ASSOCIATED PRESS" else True
    
    
    try:
        page_num = soup.find("span", attrs={"itemprop":"printPage"}).string
    except:
        page_num = ''
        ## NB: on some blogs, the page number is in a block of text, and we'd need to do regex extraction to get it.
        ##
        ## eg. http://well.blogs.nytimes.com/2013/11/11/women-find-orgasms-elusive-in-hookups/
        ## at the end of the actual article text there's this <h6> tag:
        ## <h6 class="metaFootnote">A version of this article appears in print on 11/12/2013, on page D1 of the NewYork edition with the headline: In Hookups, Inequality Still Reigns.</h6>
   
        
    try:
        category = soup.find("meta", attrs={"name":"CG"}).attrs["content"]
    except:
        category = ''
        
        
    try:
        sub_category = soup.find("meta", attrs={"name":"SCG"}).attrs["content"]
    except:
        sub_category = ''
        
        
    if content_type == "Blogs":
        
        id_tag = soup.find("meta", attrs={"name":"communityMD5"})
        
    else:
        
        id_tag = soup.find("meta", attrs={"itemprop": "identifier", "name":"articleid"})

        
    article_id = id_tag.attrs["content"] if (id_tag) else ''
                
## NYT only gives word_length metadata on the Article API, not on Most Popular
## we compute it later on if it doesn't show up here

    try:    
        
        word_count = j["word_count"]
        
    except KeyError:
        
        word_count = ''    
        
        
## 'include' evaluates all the exclusion criteria for an article
## ie. True = keep article, False = discard article

    bad_word_count = (not(word_count > 20 or word_count == ''))
    
    bad_genre = ( (content_genre == "Obituary (Obit)") or (content_genre == "Schedule") or (content_genre == "List") or (content_genre == "Brief") or (content_genre == "Text") or corrections or survey)

    good_article_type = ((content_type == "Blogs") or (content_type == "Article"))
    
    include = False if (not nyt_original) or bad_word_count or (not good_article_type) or bad_genre else True
  
    metadata = {
                "article_id":article_id,
                "title":title,
                "content_type":content_type,
                "genre":content_genre,
                "section_formal":section_formal,
                "section":category,
                "sub_category":sub_category,
                "author":author,
                "page_num":page_num,
                "word_count":word_count,
                "include":include
                }
    
    return metadata
```

```python
##
## get article text
##


## helper function for beautiful soup
##
def is_p_and_has_no_class(tag):
    
    return (not tag.has_attr('class')) and (tag.name=='p')



## main function
##
def get_article_text(soup, content_type, genre):

    article_body = []
        
    ## after looking through the actual article page source code
    ## we can see that this 'itemprop' attribute stores all blocks of article text
    ## so we parse out only these blocks from the whole text dump (there are lots of ads, images, menus, etc)
    
    ## text scraping requires finding the different tags that mark article body text in each section
    #  eg. slightly different with blogs & reviews than with regular news 

    if content_type == "Blogs":
    
        text = soup.findAll(is_p_and_has_no_class)

    elif genre == "Review":
        
        text = soup.findAll(itemprop="reviewBody")
        
    else:
        
        text = soup.findAll("p", attrs={"itemprop":"articleBody"})
    
    ## testing
    ## print 'this is the article text:',text
    
    return text
```

```python
def cleanup_text(text, content_type):
    
    ptags = []

    try:   
        
        ct = 0   ## this counter is for error handling
        for t in text:
            
            ct+=1
            
            try:
                
                
                for x in t.descendants:
                    
                    ad_text = re.match('(Get unlimited access)|(All print options include)', x.string)
                    
                    if (not x.name) and (not ad_text):
                
                        ptags.append(x.string.encode('utf-8'))
                
            except:
                
                ## print 'failed at:',ct    ## some blogs barf on the <p> html structure for some reason
                break
    
    ## apostrophes (') get turned into a funny French a
    ## so we sub back in ' with regex
    ## we also get rid of carriage returns and other useless whitespace 
    
        remove_a = re.compile('â')
        remove_w = re.compile('[\n\r\t]')  
        remove_comma = re.compile(',')   
        
        ptags = [ remove_a.sub(r"", p) for p in ptags ]

        ptags = [ remove_w.sub(r"", p) for p in ptags ]

        ptags = [ remove_comma.sub(r"", p) for p in ptags ]

        cleaned_text = ' '.join(ptags)
        
    except:
        #print text
        cleaned_text = 'error_text'
     
    #print text
    return cleaned_text
```

```python
## extract article body text from raw html

def extract_articles(api, jsons, target_count=0):

## the API return only gives us metadata and lead paragraph
## but we do get the URL
## so here we load each article URL and scrape the text
## then we store the actual article text bodies in a list

    articles = []
    
    popular = 1 if api == "POPULAR" else 0
    
    ## counter is for matching roughly the same number of popular and non-popular articles
    ## we do popular articles first.  so when collecting unpopular articles, we keep a running count,
    ## and when we match the number of popular articles already collected, we can stop scraping.
    counter = 0
    
## STEPS IN THIS CODE BLOCK:
##
## 1) loop through each article's json
## 2) make soup for parsing
## 3) grab metadata
## 4) scrape article text from url
## 5) clean article text
## 6) store in master list of articles

## NB: The Article and Most Popular APIs have different json trees, 
## so at times we need conditional branches to assign the correct paths to the nodes we want
    
    
    
    for json_set in jsons:
    
    #1
        for j in json_set:
         
            if not popular:
                
                if counter >= target_count: break
                else: counter += 1
            
        #2
            url = j["url"] if api == 'POPULAR' else j["web_url"]
            
            ## print url
            
            soup = make_soup(url) 
   
        #3  
            metadata = get_article_metadata(soup, j, api)

            ## print metadata["title"]
        #4
            text = get_article_text(soup, metadata["content_type"], metadata["genre"])
            
        #5                 
            cleaned_text = cleanup_text(text, metadata["content_type"])
                
        #6
            
            if (metadata["include"]):
                    
                articles.append([metadata["article_id"], 
                                 popular, 
                                 metadata["content_type"], 
                                 metadata["genre"],
                                 metadata["section_formal"], 
                                 metadata["section"],
                                 metadata["sub_category"],
                                 metadata["title"],
                                 metadata["author"],
                                 metadata["word_count"],
                                 url,
                                 metadata["page_num"],  
                                 cleaned_text])

    return articles
```

```python
def make_article_df(articles):
    
    df = pd.DataFrame(index = range(0,len(articles)), 
                              columns = ["article_id",
                                         "popular",
                                         "content_type",
                                         "genre",
                                         "section_formal",
                                         "section",
                                         "sub-category",
                                         "title",
                                         "author",
                                         "post_length",
                                         "url",
                                         "page_num",
                                         "text"])

    for idx, a in enumerate(articles):
    
        df.ix[idx, "article_id"] = a[0]
        df.ix[idx, "popular"] = a[1]
        df.ix[idx, "content_type"] = a[2]
        df.ix[idx, "genre"] = a[3]
        df.ix[idx, "section_formal"] = a[4]
        df.ix[idx, "section"] = a[5]
        df.ix[idx, "sub-category"] = a[6]
        df.ix[idx, "title"] = a[7]
        df.ix[idx, "author"] = a[8]
        df.ix[idx, "post_length"] = a[9]
        df.ix[idx, "url"] = a[10]
        df.ix[idx, "page_num"] = a[11]
        df.ix[idx, "text"] = a[12]
        
    return df
```

```python
def post_length(text):
    words_df = pd.DataFrame(text.split(' '), columns=['words'])
    words_df['words'] = words_df.words.apply(lambda x: str.lower(re.sub("\W", '', x)))
    words_df['word_length'] = words_df.words.apply(lambda x: len(x))
    words_df = words_df[words_df.word_length > 0]
    tot_word_count = len(words_df)
    
    return tot_word_count


def avg_word_length(text):
    words_df = pd.DataFrame(text.split(' '), columns=['words'])
    words_df['words'] = words_df.words.apply(lambda x: str.lower(re.sub("\W", '', x)))           
    words_df['word_length'] = words_df.words.apply(lambda x: len(x))
    words_df = words_df[words_df.word_length > 0]      
    average_length = round(np.mean(words_df.word_length.values), 2) 
    
    return average_length


def avg_senti_score(text):
    #now = time.time()
    words_df = pd.DataFrame(text.split(' '), columns=['words'])
    words_df['words'] = words_df.words.apply(lambda x: str.lower(re.sub("\W", '', x)))
    words_df['word_length'] = words_df.words.apply(lambda x: len(x))
    words_df = words_df[words_df.word_length > 0]
    words_df['in_sentiment'] = False
    words_df['in_sentiment'][words_df.words.isin(sentiment_df.word)] = True
    words_df['happiness_avg'] = np.nan
    words_df['happiness_avg'][words_df.in_sentiment == True] = words_df.words[words_df.in_sentiment == True].apply(lambda x: 
                float(sentiment_df.happiness_average[sentiment_df.word == x].values))
    average_score = round(np.mean(words_df.happiness_avg[~words_df.happiness_avg.isnull()].values), 3)
    
    return float(average_score)


def avg_senti_score_2(text):
    #now = time.time()
    words_df = pd.DataFrame(text.split(' '), columns=['words'])
    words_df['words'] = words_df.words.apply(lambda x: str.lower(re.sub("\W", '', x)))
    words_df['word_length'] = words_df.words.apply(lambda x: len(x))
    words_df = words_df[words_df.word_length > 0]
    words_df['in_sentiment'] = False
    words_df['in_sentiment'][words_df.words.isin(sentiment_df_NS.word)] = True
    words_df['happiness_avg_2'] = np.nan
    words_df['happiness_avg_2'][words_df.in_sentiment == True] = words_df.words[words_df.in_sentiment == True].apply(lambda x: 
                float(sentiment_df_NS.happiness_average[sentiment_df_NS.word == x].values))
    average_score_2 = round(np.mean(words_df.happiness_avg_2[~words_df.happiness_avg_2.isnull()].values), 3)
    
    return float(average_score_2)


def dyn_senti_score(text):    
    words_df = pd.DataFrame(text.split(' '), columns=['words'])
    words_df['words'] = words_df.words.apply(lambda x: str.lower(re.sub("\W", '', x)))
    words_df['word_length'] = words_df.words.apply(lambda x: len(x))
    words_df = words_df[words_df.word_length > 0]
    words_df['in_sentiment'] = False
    words_df['in_sentiment'][words_df.words.isin(sentiment_df_NS.word)] = True
    words_df['happiness_avg'] = np.nan
    words_df['happiness_avg'][words_df.in_sentiment == True] = words_df.words[words_df.in_sentiment == True].apply(lambda x: 
                float(sentiment_df_NS.happiness_average[sentiment_df_NS.word == x].values))
    avg_sent_score = round(np.mean(words_df.happiness_avg[~words_df.happiness_avg.isnull()].values), 3)
    sent_min = words_df['happiness_avg'].min(axis=0)
    sent_max = words_df['happiness_avg'].max(axis=0)
    sent_range = sent_max - sent_min
    range_constant = 0.1
    dispersion = range_constant * sent_range
    ## lower and upper boundaries for dynamic exclusion
    dyn_exclude_min = avg_sent_score - dispersion
    dyn_exclude_max = avg_sent_score + dispersion
    words_df['dyn_exclusion'] = 'True'
    words_df['dyn_exclusion'][(words_df.happiness_avg >=dyn_exclude_max)] = 'False'
    words_df['dyn_exclusion'][(words_df.happiness_avg <=dyn_exclude_min)] = 'False'
    avghdyn = np.nan
    avghdyn = words_df[words_df['dyn_exclusion'] == 'False']['happiness_avg'].mean()
    
    return float(avghdyn)


def google_rank(text):   
    words_df = pd.DataFrame(text.split(' '), columns=['words'])
    words_df['words'] = words_df.words.apply(lambda x: str.lower(re.sub("\W", '', x)))
    words_df['word_length'] = words_df.words.apply(lambda x: len(x))
    words_df = words_df[words_df.word_length > 0]
    words_df['in_sentiment'] = False
    words_df['in_sentiment'][words_df.words.isin(sentiment_df_NS.word)] = True
    words_df['google_rank'] = np.nan
    words_df['google_rank'][words_df.in_sentiment == True] = words_df.words[words_df.in_sentiment == True].apply(lambda x: 
        float(sentiment_df.google_rank[sentiment_df.word == x].values))
    google_rank = round(np.mean(words_df.google_rank[~words_df.google_rank.isnull()].values), 3)
    
    return float(google_rank)


def twitter_rank(text):  
    words_df = pd.DataFrame(text.split(' '), columns=['words'])
    words_df['words'] = words_df.words.apply(lambda x: str.lower(re.sub("\W", '', x)))
    words_df['word_length'] = words_df.words.apply(lambda x: len(x))
    words_df = words_df[words_df.word_length > 0]
    words_df['in_sentiment'] = False
    words_df['in_sentiment'][words_df.words.isin(sentiment_df_NS.word)] = True
    words_df['twitter_rank'] = np.nan
    words_df['twitter_rank'][words_df.in_sentiment == True] = words_df.words[words_df.in_sentiment == True].apply(lambda x: 
        float(sentiment_df.twitter_rank[sentiment_df.word == x].values))
    twitter_rank = round(np.mean(words_df.twitter_rank[~words_df.twitter_rank.isnull()].values), 3)
    
    return float(twitter_rank)


def nyt_rank(text):   
    words_df = pd.DataFrame(text.split(' '), columns=['words'])
    words_df['words'] = words_df.words.apply(lambda x: str.lower(re.sub("\W", '', x)))
    words_df['word_length'] = words_df.words.apply(lambda x: len(x))
    words_df = words_df[words_df.word_length > 0]
    words_df['in_sentiment'] = False
    words_df['in_sentiment'][words_df.words.isin(sentiment_df_NS.word)] = True
    words_df['nyt_rank'] = np.nan
    words_df['nyt_rank'][words_df.in_sentiment == True] = words_df.words[words_df.in_sentiment == True].apply(lambda x: 
        float(sentiment_df.nyt_rank[sentiment_df.word == x].values))
    nyt_rank = round(np.mean(words_df.nyt_rank[~words_df.nyt_rank.isnull()].values), 3)
    
    return float(nyt_rank)


def lyrics_rank(text):   
    words_df = pd.DataFrame(text.split(' '), columns=['words'])
    words_df['words'] = words_df.words.apply(lambda x: str.lower(re.sub("\W", '', x)))
    words_df['word_length'] = words_df.words.apply(lambda x: len(x))
    words_df = words_df[words_df.word_length > 0]
    words_df['in_sentiment'] = False
    words_df['in_sentiment'][words_df.words.isin(sentiment_df_NS.word)] = True
    words_df['lyrics_rank'] = np.nan
    words_df['lyrics_rank'][words_df.in_sentiment == True] = words_df.words[words_df.in_sentiment == True].apply(lambda x: 
        float(sentiment_df.lyrics_rank[sentiment_df.word == x].values))
    lyrics_rank = round(np.mean(words_df.lyrics_rank[~words_df.lyrics_rank.isnull()].values), 3)
    
    return float(lyrics_rank)


def num_sentence(text):
    return len(token.sent_tokenize(text))
    
    
def lexical_density(text):
    keep = []
    tokenized = token.word_tokenize(text)
    for word in tokenized:
        if len(word) > 5:
            keep.append(word)
    lexden = float(len(keep))/float(len(tokenized))
    
    return lexden
    
    
def words_per_sentence(text):
    tokenize = token.sent_tokenize(text)
    numsentences = len(tokenize)
    wordsinsentence = []
    for sentence in tokenize:
        wordsinsentence.append(len(token.word_tokenize(sentence)))
        
    return sum(wordsinsentence)/numsentences


d = cmudict.dict() 

def nsyl(word): 
  return [len(list(y for y in x if isdigit.y[-1])) for x in d[word.lower()]]



def FK_GradeLevel(text):
    #now = time.time()
    #.39 * (words_per_sentence(text)) + 11.8 * (
    tokenize = token.word_tokenize(text)
    syllist = []
    for word in tokenize:
        try:
            if len(nsyl(word)) > 1:
                syllist.append(nsyl(word)[0])
            else:
                syllist.append(nsyl(word))
        except:
            syllist.append(0)
            
    syllables = float(sum([np.array(i) for i in syllist]))
    
    score = .39 * (words_per_sentence(text)) + 11.8 * (syllables/post_length(text)) - 15.59
    
    return abs(score)

def syllablesComp(text):
    #.39 * (words_per_sentence(text)) + 11.8 * (
    tokenize = token.word_tokenize(text)
    syllist = []
    for word in tokenize:
        try:
            if len(nsyl(word)) > 1:
                syllist.append(nsyl(word)[0])
            else:
                syllist.append(nsyl(word))
        except:
            syllist.append(0)
            
    syllablesI = float(sum([np.array(i) for i in syllist if i > 4]))
    
    return syllablesComp

def FogIndex(text):
    score = .4 * (words_per_sentence(text)) + 100 * (syllablesI(text)/post_length(text))
    
    return score 
```

#### Make XY Function

This function takes in the prepared dataframe source and outputs X (an array of features) and Y (an array of known scores).

```python
# defining the make_xy() function
# "rating_col" parameter is the name of the column where the ratings are;
# can be continuous or binary values
# features parameter is a list of the feature functions
def make_xy(prep_data, rating_col, features):
    
    # using copy of dataframe just in case
    prep_copy = prep_data.copy()
    
    # converting relevant columns to floats
    
    for func in features:
        #print func
        prep_copy[func.__name__] = prep_copy[func.__name__].apply(lambda x: float(x))
        prep_copy = prep_copy[~prep_copy[func.__name__].isnull()].reset_index(drop = True)
    
    feature_array = []
    for i in xrange(len(prep_copy)):
        feature_list = [float(prep_copy[func.__name__][prep_copy.index == i].values[0]) for func in features]
        feature_array.append(feature_list)
    
    # making 2D array of features
    x = np.array(feature_array)
    
    # renaming the column that contains the ratings
    prep_copy = prep_copy.rename(columns={rating_col: 'rating_col'})
    
    # making numpy array of the naive happy values we made up
    y = np.array(prep_copy.rating_col.values)
    
    # returning x and y in a tuple
    return x, y
```

##### 5.1.4 Select Model
Our basic assumption is the interdependency of each feature within the feature set. As we have only got experience with this naive assumption, we concentrated in this direction. As the characteristics of our features are continous values the classic Multinomial Naive Bayesian (NB) and also the Bernoulli based approach cannot be used. The gaussian NB model is the only naive bayes model capable of dealing with continous feature sets. Alternative models have been tested and showed similar outcomes based on the same feature set. Information based on: http://scikit-learn.org/stable/modules/naive_bayes.html

Here we define a function that measures the accuracy of our model.

```python
def accuracy_func(classifier, x, y):

    clf = classifier
    predict_y = clf.predict(x)
    actual_y = y
    
    diff = abs(actual_y - predict_y)
    
    emprange = abs(max(actual_y) - min(actual_y))
    
    accuracy2 = (100. - float(np.mean(diff, dtype=float) / float(emprange) * 100.))
    
    return accuracy2
    
```

Here we define a function that allows us to cross validate the accuracy of our model.

```python
def cv_score(classifier, x, y, score_func, n_fold=5):
    clf = classifier
    result = 0
    nfold = n_fold
    for train, test in KFold(y.size, nfold): # split data into train/test groups, 5 times
        clf.fit(x[train], y[train]) # fit
        result += score_func(clf, x[test], y[test]) # evaluate score function on held-out data
    return result / nfold
```

Here we define an accuracy plot method that will enable visualization of the accuracy of any model we use to train and test our datasets.

```python
# a function that produces an accuracy plot
# "clf" is the classifier to use
# "clf_title" is just the STRING name of the classifier
# (x, y) is a tuple of the x and y made from make_xy
# the last parameter can be substituted with train_test_split(x, y)

def accuracy_plot(classifier, clf_title, (x, y), (x_train, y_train, x_test, y_test)):
    
    #create classifier (place to change parameters)
    clf = classifier
    # fit model
    clf.fit(x_train, y_train)
     
    #predict testdata
    predictedtest = clf.predict(x_test)
    #predict traindata
    predictedtrain = clf.predict(x_train)
     
    #get length of current testset
    print 'Lenght of actual value array:', len(y_test)
    #get length of current trainset
    print 'Length of predicted value array', len(predictedtest)
    print 
        
    #get numerated x for each sampleitem
    xscale = list(np.arange(len(y_test)))
     
    #create temporary df to be able to sort (this case test data)
    tempdf = pd.DataFrame(y_test, columns = ['Ytest'])
    tempdf['prediction'] = predictedtest
    tempdf = tempdf.sort('Ytest').reset_index(drop = True)
    #tempdf = tempdf.sort('prediction').reset_index(drop = True)
     
    #set x for graph after sorting
    tempdf['xscale'] = xscale
     
    #print head of tempdf
    #print tempdf.head()
     
    #calculate accuracy for test data
    predicted = predictedtest
    actual = y_test
    difference = abs(actual-predicted)
    #set range scale manually!!
    rangeval = abs(1 - 9)
    #set range scale manually!!
    emprange = abs(max(y_test) - min(y_test))
    #accuracy = (100. - float(np.mean(difference, dtype=float) / float(rangeval) * 100.))
    accuracy2 = (100. - float(np.mean(difference, dtype=float) / float(emprange) * 100.))
    print 'Test data accuracy using empirical range:', accuracy2
     
    #calculate accuracy for train data
    predicted = predictedtrain
    actual = y_train
    difference = abs(actual-predicted)
    #set range scale manually!!
    rangeval = abs(1 - 9)
    #set range scale manually!!
    emprange = abs(max(y_train) - min(y_train))
    #accuracy = (100. - float(np.mean(difference) / float(rangeval) * 100.))
    accuracy2 = (100. - float(np.mean(difference, dtype=float) / float(emprange) * 100.))
    print 'Train data accuracy using empirical range (should be higher):', accuracy2
    print
     
    # calculating cross-validated accuracy
    cv_accuracy = cv_score(clf, x, y, accuracy_func)
    print 'Cross-validated accuracy:', cv_accuracy
    print
        
    #print currently used feature functions
    print 'Used feature functions:\n', [i.__name__ for i in features]
    print
        
    #set canvas size
    plt.figure(figsize=(15,10))
    #set graph dims
    plt.xlim(-1,len(xscale)+1)
    plt.ylim(min(y_test-1), max(y_test)+1)
     
    line = np.polyfit(xscale,tempdf.prediction,1)
    #convert into function
    func = np.poly1d(line)
    #set trendlinetext for legend (retrieves function)
    trendlinetxt = 'Trendline: y = %sx + %s' % (str(round(line[0],10)), str(round(line[1],10)))
    #print trendline
    plt.plot(xscale, func(xscale), 'k', color = 'green', linewidth=15, label=trendlinetxt)
     
    #in this constellation the predicted values will be plotted in front of the predicted ones
    #plt.plot(tempdf.xscale, tempdf.prediction, 'bo')
    plt.plot(tempdf.xscale, tempdf.Ytest, 'ro', label='Actual')
    plt.plot(tempdf.xscale, tempdf.prediction, 'bo', label='Predicted')
    plt.xlabel('\n\nTexts')
    plt.ylabel('Sentiment Score\n')
    plt.title(clf_title + '\n')
    plt.legend(loc='best')
    plt.show()
 
```
