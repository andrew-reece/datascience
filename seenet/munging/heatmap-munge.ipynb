We do this so we can normalize the actual pair frequencies later on in the time series
<br />Otherwise, if there are a lot of, say, LIB2 in the dataset, then LIB2-LIB2 pairings will be higher simply because of their greater total population, which would give them the appearance of a more important pairing.  
<br />Dividing by the total number of possible such pairings keeps everything as a proportion, rather than a raw, unstandardized value.

```python
# make_time_array()
# modified from andrew-munging-2...here it hard codes start/end dates

def make_time_array(period):
    
    mintime = datetime.date(2008, 10, 1)
    maxtime = datetime.date(2009, 6, 24)
    jump = datetime.timedelta(days=period) # sets time jump to period argument
    
    # create array of two-week jumps from study start to study end
    time_array = []
    current_time = mintime
    while current_time <= maxtime:
        time_array.append(current_time)
        current_time += jump

    return time_array
```

```python
# attr_df = df of subject attributes (politics, music, etc)
# series_df = df of time series variable (proximity or communication)
# times = array of time intervals across time series data

def make_pairwise_attr_groups(attr_df, attr, series_df, times):

    cols = range(0,len(times))
    time_series = []
    
    for idx in cols:
        
        # colnames of series_df are datetime strings, so we format them from times array
        colname = datetime.datetime.strftime(times[idx],"%Y-%m-%d")
        prev_colname = datetime.datetime.strftime(times[idx-1],"%Y-%m-%d")
        
        # if first iteration, then there won't be a previous column...
        prev_val = series_df[prev_colname] if idx > 0 else 0
        
        # only get pairwise data which changed from previous timepoint
        updated = series_df.ix[series_df[colname] > prev_val, ["source","target"]]
        
        pairs = []
        for idx, row in updated.iterrows():
            user1 = row[0]
            user2 = row[1]
            # get libcon status for each user
            user1_attr = attr_df.ix[attr_df.user_id == user1, attr].values
            user2_attr = attr_df.ix[attr_df.user_id == user2, attr].values
            
            # if both users have attr values, then enter as instance of pairwise libcon occur.
            if (len(user1_attr) > 0) and (len(user2_attr) > 0):
                if (str(user1_attr[0]) != "nan") and (str(user2_attr[0]) != "nan"):
                    sorted_vals = sorted([user1_attr[0],user2_attr[0]])
                    pairwise = ''.join([str(sorted_vals[0]),"-",str(sorted_vals[1])])
                    pairs.append(pairwise)
                    # we have a symmetric matrix, so add reversed pair as well
                    pairwise_reverse = ''.join([str(sorted_vals[1]),"-",str(sorted_vals[0])])
                    pairs.append(pairwise_reverse)
                
        df = pd.DataFrame(pairs)
        colname = ''.join([attr,'_pairs'])
        df.columns = [colname]
        time_series.append(df.groupby(colname).groups)
        
    return time_series
```

```python
# makes heatmap matrix
# indices = pairwise category combos
# columns = time intervals across time series

def make_heatmap(series, cats, colnames, baserates):
    filler = np.zeros((len(baserates.pairs),len(colnames)))
    heatmap = pd.DataFrame(columns = colnames, index = baserates.pairs, data=filler)
    heatmap["total"] = 0
    for idx, col in enumerate(series):
        running_total = 0
        for key, val in col.iteritems():
            #print 'key'
            #print key
            heatmap.ix[key,"total"] += float(len(val))/float(baserates[baserates.pairs==key]["total"].values[0])
            heatmap.ix[key,idx] = float(len(val))/float(baserates[baserates.pairs==key]["total"].values[0]) 
    return heatmap.reset_index()
```

```python
# sdata: subject data
# tsdata: time series data
# var: variable for heatmap

def heatmap_wrapper(sdata, tsname, tsdata, var, var_range):

    print ''.join(["LOADING: ",tsname," _ ",var])
    
# get baserates    
    baserates = get_baserates(sdata, var, var_range)    
# get list of all individual values
    all_vals = np.repeat( baserates.keys(), baserates.values(), axis=0)
    #print baserates.keys()
    #print baserates.values()
# get all combinations of values (with replacement), store as df
    all_pairs = pd.DataFrame([''.join([str(pair[0]),'-',str(pair[1])]) for pair in it.product(all_vals, repeat=2)])
    all_pairs.columns = ["pairs"]
    #all_pairs.sort(column="pairs", axis=0, ascending=True, inplace=True)
    #print all_pairs.ix[:,"pairs"].unique()
# get count of each pairing
    pair_cts = all_pairs.groupby("pairs").apply(lambda x: len(x)).reset_index()
    pair_cts.columns = ["pairs","total"]
    #print pair_cts["pairs"].unique()
    time_array = make_time_array(14)
    
    time_series = make_pairwise_attr_groups(sdata, var, tsdata, time_array)
    
    heatmap = make_heatmap(time_series, var_range, time_array, pair_cts)
    
# save to csv
    filename = ''.join(["../data/",var,"-",tsname,"-heatmap.csv"])
    heatmap.to_csv(filename, float_format="%.4f", index_label="pairname")

# float_format is used here as otherwise values get a decimal placeholder
# index_label gives the index column a header name


```
